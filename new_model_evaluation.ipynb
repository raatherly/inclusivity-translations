```python
# prompt: import an xlsx file

import pandas as pd

# Replace 'your_file.xlsx' with the actual path to your Excel file
# If the file is in your current working directory, you can just use the filename
file_path = '/content/df_fr_fri_test.xlsx'

try:
  dataset = pd.read_excel(file_path)
  print(dataset.head()) # Print the first few rows to verify
except FileNotFoundError:
  print(f"Error: File '{file_path}' not found.")
except Exception as e:
  print(f"An error occurred: {e}")
```

```python
# Prepare references and predictions
# The lambda function iterates over the entire thing and put it into a list format
# The metrics expect the references to be a list of lists,
# even if there’s only one reference sentence per prediction.
# BLEU and ROUGE expect the predictions to be a simple list (not a list of lists).
references = dataset['fri'].apply(lambda x: [x]).tolist()
predictions = dataset['fr_inclusive_predicted'].tolist()
```

```python
import evaluate

# Load evaluation metrics
bleu = evaluate.load("bleu")
rouge = evaluate.load("rouge")

# Compute BLEU
bleu_score = bleu.compute(predictions=predictions, references=references)

# Compute ROUGE
rouge_score = rouge.compute(predictions=predictions, references=references)

print("BLEU:", bleu_score)
print("ROUGE:", rouge_score)
```

```python
# Check for any NaN values or unexpected types
print("Any NaNs in 'fr' column:", dataset['fr'].isna().sum())
print("Any NaNs in 'fr_inclusive_predicted' column:", dataset['fr_inclusive_predicted'].isna().sum())
print("Any NaNs in 'fri' column:", dataset['fri'].isna().sum())

# Check the data types in the columns
print("Data types of 'fr' column:", dataset['fr'].dtype)
print("Data types of 'fr_inclusive_predicted' column:", dataset['fr_inclusive_predicted'].dtype)
print("Data types of 'fri' column:", dataset['fri'].dtype)
```

```python
# Check for any rows where 'fr' or 'fr_inclusive_predicted' or 'fri' are floats or other unexpected types
print(dataset[dataset['fr'].apply(lambda x: isinstance(x, float))].head())
print(dataset[dataset['fr_inclusive_predicted'].apply(lambda x: isinstance(x, float))].head())
print(dataset[dataset['fri'].apply(lambda x: isinstance(x, float))].head())
```

```python
print(f"Rows before cleaning: {dataset.shape[0]}")

# Drop rows where any of the relevant columns have NaN values
dataset_clean = dataset.dropna(subset=['fr', 'fr_inclusive_predicted', 'fri'])

# Verify that the NaNs were removed
print(f"Rows after cleaning: {dataset_clean.shape[0]}")
```

```python
from comet import download_model, load_from_checkpoint

# Load a pre-trained COMET model
comet_model_path = download_model("Unbabel/wmt20-comet-da")
comet_model = load_from_checkpoint(comet_model_path)

# Prepare data
'''
COMET requires a source, reference, and hypothesis (prediction)
It assesses fluency sound of the translations

Sources (src): original French sentences
Predictions (mt):  inclusive French translations produced by the model
References (ref): standard French translations
'''
# Prepare data
comet_data = [{"src": src, "mt": pred, "ref": ref[0]} for src, pred, ref in zip(dataset_clean['fr'], dataset_clean['fr_inclusive_predicted'], dataset_clean['fri'].apply(lambda x: [x]))]

# Compute COMET score
comet_score = comet_model.predict(comet_data)

print("COMET Score:", comet_score)
```
